```python


file_path = "share_indian_tonnage_age_wise_31_dec2022.csv"
df = pd.read_csv(file_path)


sns.set_theme(style="whitegrid")


fig, axes = plt.subplots(5, 4, figsize=(20, 20))
axes = axes.flatten()

columns = df.columns[2:]
colors = ["blue", "green", "red", "purple"]


for i in range(20):
    col_idx = i % len(columns)
    region = "Coastal" if i < 10 else "Overseas"
    subset = df[df["Region"] == region]
    
    sns.lineplot(ax=axes[i], x=subset["Year"], y=subset[columns[col_idx]], color=colors[col_idx])
    axes[i].set_title(f"{columns[col_idx]} ({region})")
    axes[i].set_xlabel("Year")
    axes[i].set_ylabel("Percentage")

plt.tight_layout()
plt.show()
```
